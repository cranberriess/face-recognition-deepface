```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
!wget model structure: https://github.com/opencv/opencv/raw/3.4.0/samples/dnn/face_detector/deploy.prototxt
!wget pre-trained weights: https://github.com/opencv/opencv_3rdparty/raw/dnn_samples_face_detector_20170830/res10_300x300_ssd_iter_140000.caffemodel
```

```python
detector = cv2.dnn.readNetFromCaffe("deploy.prototxt" , "res10_300x300_ssd_iter_140000.caffemodel")
```

```python
image = cv2.imread('/content/drive/MyDrive/i_1.jpg')
base_img = image.copy()
original_size = base_img.shape
target_size = (300, 300)
image = cv2.resize(image, target_size)
aspect_ratio_x = (original_size[1] / target_size[1])
aspect_ratio_y = (original_size[0] / target_size[0])
```

```python
imageBlob = cv2.dnn.blobFromImage(image = image)
```

```python
detector.setInput(imageBlob)
```

```python
detections = detector.forward()
```

```python
detections[0][0].shape
```

```python
detections_df = pd.DataFrame(detections[0][0], columns = ["img_id", "is_face", "confidence", "left", "top", "right", "bottom"])
```

```python
detections_df = detections_df[detections_df['is_face'] == 1] #0: background, 1: face
```

```python
detections_df = detections_df[detections_df['confidence'] >= 0.90]
```

```python
from deepface.basemodels import Facenet
from deepface.commons import functions

import matplotlib.pyplot as plt
import numpy as np
```

```python
model = Facenet.loadModel()
```

```python
try:
	input_shape = model.layers[1].input_shape[1:3]
except: #issue 470
	input_shape = model.layers[0].input_shape[0][1:3]

print("model input shape: ", model.layers[0].input_shape[1:])
print("model output shape: ", model.layers[-1].input_shape[-1])
```

```python
#img1 = functions.detectFace('/content/drive/MyDrive/i.jpg', input_shape)
img1 = functions.preprocess_face('/content/drive/MyDrive/i.jpg', input_shape)
img1_representation = model.predict(img1)[0,:]
```

```python
img2 = functions.preprocess_face('/content/drive/MyDrive/i_1.jpg', input_shape)
img2_representation = model.predict(img2)[0,:]
```

```python
distance_vector = np.square(img1_representation - img2_representation)
#print(distance_vector)

distance = np.sqrt(distance_vector.sum())
print("Euclidean distance: ",distance)
```

```python
img1_graph = []; img2_graph = []; distance_graph = []

for i in range(0, 200):
	img1_graph.append(img1_representation)
	img2_graph.append(img2_representation)
	distance_graph.append(distance_vector)

img1_graph = np.array(img1_graph)
img2_graph = np.array(img2_graph)
distance_graph = np.array(distance_graph)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(3,2,1)
plt.imshow(img1[0][:,:,::-1])
plt.axis('off')

ax2 = fig.add_subplot(3,2,2)
im = plt.imshow(img1_graph, interpolation='nearest', cmap=plt.cm.ocean)
plt.colorbar()

ax3 = fig.add_subplot(3,2,3)
plt.imshow(img2[0][:,:,::-1])
plt.axis('off')

ax4 = fig.add_subplot(3,2,4)
im = plt.imshow(img2_graph, interpolation='nearest', cmap=plt.cm.ocean)
plt.colorbar()

ax5 = fig.add_subplot(3,2,5)
plt.text(0.35, 0, "Distance: %s" % (distance))
plt.axis('off')

ax6 = fig.add_subplot(3,2,6)
im = plt.imshow(distance_graph, interpolation='nearest', cmap=plt.cm.ocean)
plt.colorbar()

plt.show()
```
